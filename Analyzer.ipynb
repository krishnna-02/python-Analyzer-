```python
import numpy as np

array = None

while True:
    print("\nWelcome to the NumPy Analyzer!")
    print("=" * 40)
    print("Choose an option:")
    print("1. Create a Numpy Array")
    print("2. Perform Mathematical Operations")
    print("3. Combine or Split Arrays")
    print("4. Search, Sort, or Filter Arrays")
    print("5. Compute Aggregates and Statistics")
    print("6. Exit")

    choice = int(input("Enter your choice: "))

    
    
    if choice == 1:
        print("\nArray Creation:")
        print("Select the type of array to create:")
        print("1. 1D Array")
        print("2. 2D Array")
        print("3. 3D Array")

        arr_type = int(input("Enter your choice: "))

        if arr_type == 2:
            rows = int(input("\nEnter the number of rows: "))
            cols = int(input("Enter the number of columns: "))
            elements = list(map(int, input(
                f"Enter {rows * cols} elements for the array separated by space: "
            ).split()))

            array = np.array(elements).reshape(rows, cols)
            print("\nArray created successfully:")
            print(array)


            
            print("\nChoose an operation:")
            print("1. Indexing")
            print("2. Slicing")
            print("3. Go Back")

            op = int(input("Enter your choice: "))

            if op == 2:
                r = input("\nEnter the row range (start:end): ")
                c = input("Enter the column range (start:end): ")

                rs, re = map(int, r.split(":"))
                cs, ce = map(int, c.split(":"))

                print("\nSliced Array:")
                print(array[rs:re, cs:ce])


    
    elif choice == 2:
        print("\nMathematical Operations:")
        print("Choose a mathematical operation:")
        print("1. Addition")
        print("2. Subtraction")
        print("3. Multiplication")
        print("4. Division")

        op = int(input("Enter your choice: "))

        elements = list(map(int, input(
            "Enter the same-size array elements (6 elements separated by space): "
        ).split()))

        second_array = np.array(elements).reshape(array.shape)

        print("\nOriginal Array:")
        print(array)
        print("\nSecond Array:")
        print(second_array)

        if op == 1:
            print("\nResult of Addition:")
            print(array + second_array)


    
    elif choice == 3:
        print("\nChoose an option:")
        print("1. Combine Arrays")
        print("2. Split Array")

        op = int(input("Enter your choice: "))

        if op == 1:
            elements = list(map(int, input(
                "\nEnter the elements of another array to combine (6 elements separated by space): ").split()))

            second_array = np.array(elements).reshape(array.shape)

            print("\nOriginal Array:")
            print(array)
            print("\nSecond Array:")
            print(second_array)

            combined = np.vstack((array, second_array))
            print("\nCombined Array (Vertical Stack):")
            print(combined)


    
    elif choice == 4:
        print("\nChoose an option:")
        print("1. Search a value")
        print("2. Sort the array")
        print("3. Filter values")

        op = int(input("Enter your choice: "))

        if op == 2:
            print("\nOriginal Array:")
            print(array)

            sorted_array = np.sort(array, axis=1)
            print("\nSorted Array:")
            print(sorted_array)
            print("(Sorting applied row-wise.)")


    
    elif choice == 5:
        print("\nAggregates and Statistics:")
        print("Choose an aggregate/statistical operation:")
        print("1. Sum")
        print("2. Mean")
        print("3. Median")
        print("4. Standard Deviation")
        print("5. Variance")

        op = int(input("Enter your choice: "))

        print("\nOriginal Array:")
        print(array)

        if op == 3:
            print("\nMedian of Array:", np.median(array))


    
    elif choice == 6:
        print("\nThank you for using the NumPy Analyzer! Goodbye!")
        break
```
